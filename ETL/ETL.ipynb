```python
import pandas as pd
```

```python
# Cargar el DataFrame
df = pd.read_excel('Terrazas_202104.xlsx')

# Verificar cuántos valores nulos tiene cada columna
print(df.isnull().sum())
# Inspeccionar cuántos valores nulos tiene cada fila
df.shape

# Mostrar el DataFrame original
print("DataFrame original:")
print(df)

# Calcular el umbral de columnas no nulas
threshold = len(df.columns) / 2

# Filtrar los registros que contengan valores nulos en más del 50% de sus columnas
filtered_df = df.dropna(thresh=threshold+1)

# Calcular el número de registros eliminados
registros_eliminados = len(df) - len(filtered_df)

# Mostrar el DataFrame filtrado y el número de registros eliminados
print("\nDataFrame filtrado:")
print(filtered_df)
print(f"\nNúmero de registros eliminados: {registros_eliminados}")
```

```python
# Convertir las columnas que empiecen por "hora" a formato de hora
hora_columns = [col for col in df.columns if col.startswith('hora')]
for col in hora_columns:
    df[col] = pd.to_datetime(df[col], format='%H:%M:%S').dt.time

# Mostrar el DataFrame transformado
print("\nDataFrame transformado:")
print(df)

# Imprimir el tipo de dato de cada una de las columnas
print("Tipos de datos de cada columna:")
print(df.dtypes)
```

```python
# Crear una nueva columna que calcule el ratio entre "Superficie_TO" y "id_terraza"
df['Superficie_TO'] =  df['Superficie_ES'] + df['Superficie_RA']
df['Ratio'] = df['Superficie_TO'] / df['id_terraza']

print(df['Ratio'])
```
